# **Resample Stock Data**

```python
dr = yf.download('DRREDDY.BO', period='1wk', interval='1m')
```

```python
dr
```

```python
y=dr.resample('10T').mean()
y
```

```python
# Plotting the 10minute stock data
y.Close.plot(figsize=(11,4))
```

```python
r=y.resample('15T').sum()
r
```

```python
# Plotting the 15minutes data
r.Close.plot(figsize=(11, 4))
```

```python
# Plotting the original 'Close' stock data
dr.Close.plot(figsize=(11,4))
```

# Revenues: fetching the revenues for Dr.Reddy Laboratories
